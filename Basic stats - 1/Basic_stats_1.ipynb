```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sales_data_with_discounts.csv')
df.head()
```

```python
df.info()
```

```python
num_cols = ['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']

print("Means :")
for col in num_cols:
    print(f'{col} : {np.round(df[col].mean(),2)}')
```

```python
print("Median :")
for col in num_cols:
    print(f'{col} : {np.round(df[col].median(),2)}')
```

```python
print("Mode :")
for col in num_cols:
    print(f'{col} : {np.round(df[col].mode(),2)}')
```

```python
print("standard deviation :")
for col in num_cols:
    print(f'{col} : {np.round(df[col].std(),2)}')
```

Observations :
i)The dataset highlights that Mobiles generate the highest revenue due to their high prices, despite varying in volumes.
ii)Discounts significantly impact Net Sales Value across all categories
iii)With Lifestyle products often having the highest discount rates. 
iv)FMCG products show higher volumes but lower revenue due to their low price. 
v)Daily sales fluctuations suggest varying consumer demand, with no strong day-specific trends.

```python
# histogram
plt.figure(figsize=(12,7))
for i,col in enumerate(num_cols,1):
    plt.subplot(2,3,i)
    plt.hist(df[col])
    plt.title(f'histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('frequency')

plt.tight_layout()
```

```python
# box plot
plt.figure(figsize=(12,7))
for i,col in enumerate(num_cols,1):
    plt.subplot(2,3,i)
    plt.boxplot(df[col])
    plt.title(f'boxplot for {col}')
    plt.xlabel(col)
    plt.ylabel('count')

plt.tight_layout()
```

```python
# bar graph

cat_cols = ['Date','Day','SKU','Volume','BU','Brand','Model']
plt.figure(figsize=(10,20))

for i,col in enumerate(cat_cols[:]):
    plt.subplot(4,2,i+1)
    counts = df[col].value_counts()
    plt.bar(counts.index.astype(str),counts.values)
    plt.xticks(rotation=45, ha='right')
    plt.title(f'bar graph for {col}')

plt.tight_layout()
```

Observations :
i)The right-skewed distributions of Volume, Total Sales Value, and Net Sales Value highlight that a few high-volume or high-value transactions (mostly Mobiles and some FMCG) drive revenue, while most sales are low-volume or low-value. Outliers in Volume (ex: F01) and Avg Price (ex:M10) indicate specific products with exceptional demand or pricing.
ii)Discounts vary by category, with Lifestyle and FMCG often having higher rates to boost sales, while Mobiles have lower rates to maintain premium value. Outliers in discount rates suggest targeted promotions.
iii)The bar charts equal transaction distribution hides real-world differences. FMCGs high volumes and low prices suggest mass-market appeal, while Mobiles high prices drive revenue despite lower volumes. Lifestyle products balance between these extremes, with moderate prices and volumes.

```python
# standardization 
# using z-score
# z = (x-mean)/std

plt.boxplot(df[num_cols])
plt.title("--Before Standardization--")
plt.show()
```

```python
print("z = (x-mean)/std")
df_z = df.copy()

# Apply z-score normalization
for col in num_cols:
    mean = df[col].mean()
    std = df[col].std()
    df_z[col] = (df[col] - mean) / std
```

```python
plt.figure(figsize=(5,15))
df_z.plot(kind='box')
plt.xticks(rotation=45, ha='right')
plt.title("--After Standardization--")
plt.show()
```

-Standardization and one-hot encoding are preprocessing steps that transform data into a format suitable for machine learning. 
-Standardization ensures numerical features contribute fairly and models converge efficiently by removing 'units' scale & converting them into mean = 0 & std = 1
-Encoding :One-hot encoding transforms categorical data into binary vectors (1s and 0s) based on the presence of categories.
-These steps would allow accurate modeling of sales patterns, revealing insights mentioned above.
-Neglecting these steps risks poor model performance, biased results, and missed opportunities for actionable insights.
